**表示层**
是用来与之交互的部分， 也就是所谓的用户界面 

**业务层**
这一层负责特定程序的逻辑。 如D4单元格定义了金额如何被还转成美金的

**数据层**
这一层负责访问数据。 D4单元格的VLOOKUP部分负责完成这项工作。

在专业的excel开发者会建议， 将这个excel分成多层（输入层、计算层、输出层）， 这三层单独分配一张工作表。来应对超大量的数据时候。


**Excel原则--DRY原则**

don't repeat yourself.
没有重复的代码意味着代码行数更好， 但是在excel业务中， 往往公式吴系在单元格中， 那么就会很难遵守DRY原则，这样没测启动一个新的Excel项目时， 最常见的做法就是从一个之前项目或者模板中把这些含有公式的工作部复制过来。

**测试**

通常的开发语言都会有多种单元测试的架构，但是Excel并未提供， 如$4*5=20$，代码为 $c=a*b$, 我们可以验证c时候不是等于20， 但是Excel没有这种检测工具或者代码

**版本控制**

Excel不存在版本控制， 其是Excel将代码与数据混为一起

## 现代的Excel
Power Query：用来连接各种数据源， 包括Excel工作簿、CSV文件、SQL数据库，等等。可行工程就是好处理一张工作表装不下的额数据集， 在加载数据之后，你还可以额外的操作来清理、操作数据，使分组。

Power BI： 其目的是希望通过在交互式仪表中可视化巨大的数据集使其更加容易理解, **同时BI支持python**
